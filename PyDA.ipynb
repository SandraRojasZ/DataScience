```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
data = fetch_california_housing()
```

```python
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='MedHouseValue')
```

```python
sns.boxplot(X[['MedInc', 'AveRooms', 'HouseAge']])
```

```python
sns.pairplot(pd.concat([X[['MedInc', 'AveRooms', 'HouseAge']], y], axis=1))
plt.suptitle("Relação entre variáveis e o valor da casa", y=1.02)
plt.show()
```

```python
# Normalização dos dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
```

```python
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R2: {r2:.2f}")
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=y_test, y=y_predict, alpha=0.3)
plt.xlabel("Valor Real")
plt.ylabel("Valor Previsto")
plt.title("Regressão Linear")
plt.plot([0,5], [0,5], 'r--')
plt.show()
```
