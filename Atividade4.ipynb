Utilizando uma funçõa lambda obtenha a diferença entre a média e o valor de cada subgrupo

```python
num = ['Idade', 'Registro']
```

```python
# as cidades dos funcionários
grupo_cid = df.groupby('Cidade')
filtro_cid = grupo_cid[num]
lbd = lambda x: (x-x.mean())
filtro_cid.transform(lbd)
```

```python
# a formação dos funcionários
grupo_form = df.groupby('Formação')
filtro_form = grupo_form[num]
lbd = lambda x: (x-x.mean())
filtro_form.transform(lbd)
```

```python
# o departamento dos funcionários
grupo_depa = df.groupby('Departamento')
filtro_depa = grupo_depa[num]
lbd = lambda x: (x - x.mean())
filtro_depa.transform(lbd)
```

```python
# a cidade e o departamento dos funcionários
grupo_cid_depa = df.groupby(['Cidade', 'Departamento'])
filtro_cid_depa = grupo_cid_depa[num]
lbd = lambda x: (x - x.mean())
filtro_cid_depa.transform(lbd)
```

```python
# o departamento e a formação dos funcionários
grupo_depa_form = df.groupby(['Departamento', 'Formação'])
filtro_depa_form = grupo_depa_form[num]
lbd = lambda x: (x - x.mean())
filtro_depa_form.transform(lbd)
```
